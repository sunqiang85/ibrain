## 第1章 绪论
### 1.1 什么是数理统计学
#### 1.1.1 数理统计学的任务和性质
数理统计是数学的一个分支，它是研究如何有效的收集和使用带有随机性影响的数据的一门学科
1. 有效的收集数据：全面观察、抽样调查和安排试验
2. 有效的使用数据：推断
3. 数理统计与各种专门学科的关系：共性且收随机性影响的数据收集、整理和推断问题
4. 数理统计方法的归纳性质

#### 1.1.2 数理统计学的应用
广泛应用于国家统计局、工农生产实践、计量经济学、生物医药遗传学、自然科学研究
#### 1.1.3 数理统计学的发展
两个阶段
1. 20世纪初-第二次世界大战结束：R.A. Fisher和K. Pearson为首的英国学派；还有W.S. Gosset(student), J. Neyman, E.S. Pearson(K. Pearson的儿子) A.Wald, H.Gramer
2. 二战后-现在：A.Wald的统计判决理论和Bayes学派的兴起。

### 1.2 数理统计的若干基本概念
#### 1.2.1 总体和样本
一个统计问题所研究的对象的全体称为总体（population）。在数理统计学中总体可以用一个随机变量及其概率分布来描述。
#### 1.2.2 样本空间和样本的两重性
1. 样本空间：样本$\mathbf{X}=(X_1,\ldots,X_n)$可能取值的的全体，构成样本空间，记为$\mathcal{X}$
2. 样本的两重性：抽样后是具体的数，抽样前是随机变量
3. 简单随机样本：相互独立，相同分布

#### 1.2.3 样本分布
1. 不放回抽样，样本间不是相互独立的

#### 1.2.4 统计模型
统计模型(statistical model)，就是指研究问题时所抽样本的分布。更确切的说统计模型就是样本分布族(distribution family of the sample)
1. 统计模型的确定：模型是对指定样本而言的，因此只有明确了样本的产生方法，并辅以必要的假定，才能定下模型。（比如是否有放回抽样）
2. 很多性质不一样的问题，可以归入同一模型下
3. 同一模型下可以提出很多不同的统计问题：抽样分布、参数估计和假设检验

#### 1.2.5 统计推断
1. 参数和参数空间：样本分布中位置的常数称为参数。
2. 样本分布族：样本分布，当参数取不同值时就得到不同的分布，这些样本分布就构成分布族
3. 统计推断：从总体中抽取一定大小的样本去推断总体的概率分布方法，称为统计推断（statistical inference）。统计推断问题又分为：参数统计推断（参数估计和假设检验）和非参数统计推断。统计推断内容包括：提出种种统计推断的方法；计算有关推断表示性能的数量指标；寻找或证明最优统计方法。
4. 概率论与数理统计的关系：数理统计进行统计推断的第一步是利用概率论提供的种种模型对数据建模。 

### 1.3 统计量
#### 1.3.1 统计量的定义
由样本算出的量称为统计量，或者说，统计量是样本的函数。
#### 1.3.2 若干常用的统计量
1. 样本均值
2. 样本方差
3. 样本矩
4. 二维随机向量的样本矩
5. 次序统计量及其有关的统计量：样本中位数，极值，样本p分位数，样本极差
6. 样本变异系数
7. 样本偏度
8. 样本峰度：众数附近“峰”的尖峭程度的一种度量

#### 1.3.3 经验分布函数
经验分布函数$F_n(x)$具有的大样本性质：
1. 中心极限定理
2. Bernoulli辛钦大数定律

## 第2章 抽样分布及若干预备知识
### 2.1 引言
- 样本是随机变量，具有一定的概率分布，称为样本分布（统计模型）
- 统计量是样本的函数，其分布称为抽样分布

英国统计学家R.A. Fisher曾把抽样分布、参数估计和假设检验看作统计推断的三个中心内容。

- 统计量的精确分布：当总体$X$的分布类型已知时，若对任一自然数$n$,都能导出统计量$T=T(X_1,\ldots,X_n)$分布的表达式
- 统计量的极限分布：当样本容量n趋近无穷大时，统计量的分布称为极限分布。

### 2.2 正态总体样本均值样本方差的分布

### 2.6 指数族

### 2.7 充分统计量
####  2.7.1 引言与定义
充分统计量是R.A. Fisher在其1922年的奠定性工作中提出来的。充分统计量包含了推断参数$\theta$的全部信息。
- 定义：若已知T的条件下，样本X的条件分布与参数$\theta$无关，则称$T(X)$为$\theta$的充分统计量
- 解法一：直接条件概率公式
- 解法二：随机变量矩阵变换，通过Jacobi行列式算出新随机变量的概率分布；变换后的某一随机变量（统计量）的概率分布；已知统计量后的条件概率分布

#### 2.7.2因子分解定理：
$f(x,\theta)=g(t(x),\theta)h(x)$
- 推论：T是充分统计量，$S=\phi(T)$是单值可逆函数，则S也是$\theta$的充分统计量

#### 2.7.3 极小充分统计量
- 若两个统计量T是S的函数，T=q(S)则T更简化
- 若对任意充分统计量S，都有$T=q_S(S)$，则称T是此分布族的极小充分统计量
