```python
class ColorBar:
    def __init__(self, color_num):
        self.color_num = color_num
    def get_color(self, val):
        assert val<=1 and val >=0
        color_pixel = np.array([[[round(60*val), 255,255]]], dtype= np.uint8)
        return [int(c) for c in cv2.cvtColor(color_pixel, cv2.COLOR_HSV2BGR)[0][0]]
```

```python
cbar = ColorBar(100)
canvas = np.zeros((50,1200,3), dtype=np.uint8)
for x in range(0, 1200):
    cv2.line(canvas, (x, 0), (x, 600), cbar.get_color(x/1200.0))
plt.imshow(canvas)
```

```python
plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
```
